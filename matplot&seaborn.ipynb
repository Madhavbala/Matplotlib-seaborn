```python
one = [1,2,3,4,5]
two = [1,5,4,5,6]
ones = [3,2,1,5,6]
twos = [6,2,3,4,7]
```

```python
plt.plot(one,two,'r',label = 'Loss')
plt.plot(ones,twos,'g',label = 'Profit')
plt.plot(one,two,'ro',label = 'Loss')
plt.plot(ones,twos,'go',label = 'Profit')
plt.legend()
plt.show()
```

```python
file = pd.read_csv("C:\\Users\\madhavan.bala\\Documents\\SANDY\\dataset.csv")
# print(file.head())

level = file.iloc[:,:-1].values
salary = file.iloc[:,-1].values

# print(level,salary)
plt.scatter(level,salary , color = 'red')
plt.xlabel('Salary')
plt.ylabel('Level')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
modelLR = LinearRegression()
modelLR.fit(level,salary)
```

```python
plt.scatter(level,salary,color = 'red')
plt.plot(level , modelLR.predict(level))
plt.title("Linear Regression")
plt.show()
```
